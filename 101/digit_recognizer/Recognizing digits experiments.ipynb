```python
import numpy as np
import pandas
import sklearn
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pandas.read_csv('train.csv')
```

```python
data
```

```python
labels = data['label'].values
```

```python
features = data.iloc[:,1:].values
```

```python
label_train, label_test, feat_train, feat_test = sklearn.cross_validation.train_test_split(labels, features, test_size=0.2, random_state=0)
```

```python
label_train, label_test
```

```python
feat_train, feat_test
```

```python
label_train.shape, label_test.shape, feat_train.shape, feat_test.shape
```

```python
classifier = RandomForestClassifier(n_estimators=100, n_jobs=7)
classifier.fit(feat_train, label_train)
```

```python
classifier.score(feat_train, label_train)
```

```python
classifier.score(feat_test, label_test)
```

```python
classifier = RandomForestClassifier(n_estimators=100, n_jobs=8)
classifier.fit(feat_train, label_train)
```

```python
for i in range(5):
    rf_classifier = RandomForestClassifier(n_estimators=10, n_jobs=12)
    rf_classifier.fit(feat_train, label_train)
    print(rf_classifier.score(feat_train, label_train), rf_classifier.score(feat_test, label_test))
```

```python
for i in range(5):
    rf_classifier = RandomForestClassifier(n_estimators=100, n_jobs=12)
    rf_classifier.fit(feat_train, label_train)
    print(rf_classifier.score(feat_train, label_train), rf_classifier.score(feat_test, label_test))
```

```python
for i in range(5):
    rf_classifier = RandomForestClassifier(n_estimators=1000, n_jobs=12)
    rf_classifier.fit(feat_train, label_train)
    print(rf_classifier.score(feat_train, label_train), rf_classifier.score(feat_test, label_test))
```

```python
for i in range(5):
    rf_classifier = RandomForestClassifier(n_estimators=1000, n_jobs=12, criterion='entropy')
    rf_classifier.fit(feat_train, label_train)
    print(rf_classifier.score(feat_train, label_train), rf_classifier.score(feat_test, label_test))
```
