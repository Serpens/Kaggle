```python
def get_chop_stats(img_path1, img_path2):
    img1, img2 = Image.open(img_path1), Image.open(img_path2)
    img1 = img1.resize((1024, 1024), PIL.Image.ANTIALIAS)
    img2 = img2.resize((1024, 1024), PIL.Image.ANTIALIAS)
    img1 = img1.convert('RGB')
    img2 = img2.convert('RGB')
    result = {}
    new_img = ImageChops.add(img1, img2)
    result['add'] = ImageStat.Stat(new_img)
    new_img = ImageChops.darker(img1, img2)
    result['darker'] = ImageStat.Stat(new_img)
    new_img = ImageChops.lighter(img1, img2)
    result['lighter'] = ImageStat.Stat(new_img)
    new_img = ImageChops.difference(img1, img2)
    result['difference'] = ImageStat.Stat(new_img)
    new_img = ImageChops.multiply(img1, img2)
    result['multiply'] = ImageStat.Stat(new_img)
    new_img = ImageChops.screen(img1, img2)
    result['screen'] = ImageStat.Stat(new_img)
    return result
```
